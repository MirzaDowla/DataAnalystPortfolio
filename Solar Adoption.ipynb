**Step-2 : Data Cleaning**
- Renaming the large title of postcode

```python
df1.rename(columns={'Small Unit Installation Postcode': 'Postcode'}, inplace=True)
df1['Postcode'] = df1['Postcode'].astype('string').str.zfill(4)
df1.head(2)
```

```python
# creating a separate data frame for 2001-2022
df_2001_2022 = df1[['Postcode','Previous Years (2001-2022) - Installation Quantity', 'Previous Years (2001-2022) - SGU Output Rated Output In kW',
                   'Installations Quantity Total', 'SGU Rated Output In kW Total']]

df_2001_2022.columns = ['Postcode', 'Quantity', 'kW',  'Quantity Total', 'kW Total']

df_2001_2022.head(2)
```

```python
df1.drop(columns=['Previous Years (2001-2022) - Installation Quantity', 'Previous Years (2001-2022) - SGU Output Rated Output In kW','Installations Quantity Total', 'SGU Rated Output In kW Total'], axis=1,inplace=True)
df1.head(2)
```

```python
df1.info()
```

**Step-3-: Unpivot columns**

```python
unpivoted_df1 = pd.melt(df1, id_vars=['Postcode'],var_name='Date', value_name='Value')
```

```python
unpivoted_df1
```

```python
unpivoted_df1[['Date', 'Metric']] = unpivoted_df1['Date'].str.split(' - ', expand=True)
```

```python
unpivoted_df1
```

```python
unpivoted_df1['Metric'] = unpivoted_df1['Metric'].str.split(' ').str[-1]
unpivoted_df1
```

```python
unpivoted_df1.info()
```

```python
#pivot_table = repivot on the remaining columns
unpivoted_df1_Pivot = unpivoted_df1.pivot_table(index=['Postcode', 'Date'], columns='Metric', values='Value').reset_index()
#unpivoted_df1_Pivot = unpivoted_df1.pivot_table(index=['Postcode', 'Date'], columns='Metric', values='Value')
unpivoted_df1_Pivot.columns.name = None # removes the name Metric from the index
unpivoted_df1_Pivot
```

```python
unpivoted_df1_Pivotgt0=unpivoted_df1_Pivot.loc[unpivoted_df1_Pivot['Quantity'] > 0]
unpivoted_df1_Pivotgt0
```

```python
unpivoted_df1_Pivotgt0 = unpivoted_df1_Pivotgt0.copy()
unpivoted_df1_Pivotgt0['date'] = pd.to_datetime(unpivoted_df1_Pivotgt0['Date'], format='%b %Y')
unpivoted_df1_Pivotgt0
```

**Step-4 :Preliminary Analysis**

```python
# create a data set for monthly trends!
monthly_trends_kw = unpivoted_df1_Pivotgt0.set_index('date').resample('ME')['kW'].sum().reset_index()
```

```python
sns.set_style("darkgrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_trends_kw,  marker='o', color='purple', linestyle='--', x='date', y='kW')
plt.title('Overall Monthly Trends in Added kW Capacity Across Australia')
plt.xlabel('Date')
plt.ylabel('kW Capacity Added')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**We can see below in the barplot that there is no datapoint after Jan-24, therefore there is a sudden dip in the lineplot after Jan-24**

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 5))
sns.barplot(data=monthly_trends_kw, x='date', y='kW')
plt.xlabel('Date',fontweight='bold')
plt.ylabel('SGU Rated Output In kW',fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
solar_data_2023 = unpivoted_df1_Pivotgt0[unpivoted_df1_Pivotgt0['date'].dt.year == 2023]
```

```python
postcode_capacity_2023 = solar_data_2023.groupby('Postcode')['kW'].sum().reset_index()
```

**Top Postcodes by Solar Adoption**

```python
top_10_postcodes = postcode_capacity_2023.sort_values(by='kW', ascending=False).head(10)
top_10_postcodes
```

```python
# For the top 10 postcodes, get their monthly kW added capacity
top_postcodes_monthly = solar_data_2023[solar_data_2023['Postcode'].isin(top_10_postcodes['Postcode'])]
top_postcodes_monthly = top_postcodes_monthly.groupby(['Postcode', pd.Grouper(key='date', freq='ME')])['kW'].sum().reset_index()

# Plotting the monthly kW added capacity for the top 10 postcodes
plt.figure(figsize=(14, 8))
sns.lineplot(data=top_postcodes_monthly, x='date', y='kW', hue='Postcode', palette='tab10', marker='o')
plt.title('Monthly kW Added Capacity for Top 10 Postcodes in 2023')
plt.xlabel('Date')
plt.ylabel('kW Capacity Added')
plt.legend(title='Postcode')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display the top 10 postcodes for reference
top_10_postcodes
```

```python
a=df_merged.groupby("State").agg({'kW': 'sum', 'Population': 'sum'}).reset_index()
plt.figure(figsize=(10, 6))
plt.scatter(a['Population'], a['kW'], alpha=0.7, color='green')
for i, row in a.iterrows():
    plt.text(row['Population'], row['kW'], row['State'], fontsize=9, ha='right', va='bottom')
plt.title('Population vs. Total kW Added (By State)')
plt.xlabel('Population')
plt.ylabel('Total kW Added')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Normalizing  population differences by calculating kW/10,000 people for each  state**

```python
a['kW_per_10k'] = (a['kW'] / a['Population']) * 10000
a
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(a['Population'], a['kW_per_10k'], alpha=0.7, color='green')
for i, row in a.iterrows():
    plt.text(row['Population'], row['kW_per_10k'], row['State'], fontsize=9, ha='right', va='bottom')
plt.title('Population vs. kW_per_10k Added (By State)')
plt.xlabel('Population')
plt.ylabel('kW_per_10k')
plt.grid(True)
plt.tight_layout()
plt.show()
```
